```python
import numpy as np

X = np.array([
[0,0],
[0,1],
[1,0],
[1,1]
])

def step(x):
    return 1 if x >= 0 else 0

learning_rate = 0.1
epochs = 10

weights = np.zeros(2)
bias = 0.0

Y = np.array([0,0,0,1])

for epoch in range(epochs):
    for inputs, target in zip(X,Y):
        linear_output = np.dot(weights,inputs) + bias
        prediction = step(linear_output)
        error = target - prediction
        weights += learning_rate * error * inputs
        bias += learning_rate * error

print("AND Gate")
for input_vec in X:
    result = step(np.dot(weights,input_vec) + bias)
    print(input_vec,"->",result)


weights = np.zeros(2)
bias = 0.0

Y = np.array([0,1,1,1])

for epoch in range(epochs):
    for inputs, target in zip(X,Y):
        linear_output = np.dot(weights,inputs) + bias
        prediction = step(linear_output)
        error = target - prediction
        weights += learning_rate * error * inputs
        bias += learning_rate * error

print("OR Gate")
for input_vec in X:
    result = step(np.dot(weights,input_vec) + bias)
    print(input_vec,"->",result)


weights = np.zeros(2)
bias = 0.0

Y = np.array([1,1,1,0])

for epoch in range(epochs):
    for inputs, target in zip(X,Y):
        linear_output = np.dot(weights,inputs) + bias
        prediction = step(linear_output)
        error = target - prediction
        weights += learning_rate * error * inputs
        bias += learning_rate * error

print("NAND Gate")
for input_vec in X:
    result = step(np.dot(weights,input_vec) + bias)
    print(input_vec,"->",result)


weights = np.zeros(2)
bias = 0.0

Y = np.array([1,0,0,0])

for epoch in range(epochs):
    for inputs, target in zip(X,Y):
        linear_output = np.dot(weights,inputs) + bias
        prediction = step(linear_output)
        error = target - prediction
        weights += learning_rate * error * inputs
        bias += learning_rate * error

print("NOR Gate")
for input_vec in X:
    result = step(np.dot(weights,input_vec) + bias)
    print(input_vec,"->",result)


inputs = [
[0,0],
[0,1],
[1,0],
[1,1]
]

print("XOR Gate")

for x1,x2 in inputs:

    w11 = 1
    w12 = 1
    b1 = -0.5
    h1 = step(w11*x1 + w12*x2 + b1)

    w21 = -1
    w22 = -1
    b2 = 1.5
    h2 = step(w21*x1 + w22*x2 + b2)

    w31 = 1
    w32 = 1
    b3 = -1.5
    xor = step(w31*h1 + w32*h2 + b3)

    print([x1,x2],"->",xor)


import math

def sigmoid(x):
    return 1/(1+math.exp(-x))

def dsigmoid(y):
    return y*(1-y)
```
