```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/glass.csv')
df.head(20)
```

```python
df.shape
```

```python
df['Window'] = df['Type'].map({
    1: 0, 2: 0, 3: 0, 4: 0,   # window glass
    5: 1, 6: 1, 7: 1          # non-window glass
})

df.head()
```

```python
X = df.drop(columns=['Type']).values

y = df['Window'].values

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Shuffle indices
indices = np.random.permutation(len(X))

# 80% train, 20% test
split = int(0.8 * len(X))
train_idx, test_idx = indices[:split], indices[split:]

X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y[train_idx], y[test_idx]

print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def train_logistic(X, y, lr=0.01, epochs=2000):
    m, n = X.shape
    weights = np.zeros(n)
    bias = 0

    for _ in range(epochs):
        z = np.dot(X, weights) + bias
        y_hat = sigmoid(z)

        dw = (1/m) * np.dot(X.T, (y_hat - y))
        db = (1/m) * np.sum(y_hat - y)

        weights -= lr * dw
        bias -= lr * db

    return weights, bias
```

```python
weights, bias = train_logistic(X_train, y_train)

print("Trained weights:", weights)
print("Trained bias:", bias)
```

```python
def predict(X, weights, bias):
    probs = sigmoid(np.dot(X, weights) + bias)
    return (probs >= 0.5).astype(int)
```

```python
# Predictions
y_pred = predict(X_test, weights, bias)

# Accuracy
accuracy = np.mean(y_pred == y_test)
print("Test Accuracy:", accuracy)
```
